importing libraries

```python
import json, math
from pprint import pprint
import operator
```

Defining base path

```python
root_path = "/content/gdrive/My Drive/Project_Dataset_MSVD"
```

Reading results from video frame classification model.

```python
def normalize(temp_dict):
  for key in temp_dict.keys():
    parent = temp_dict[key]
    maximum = max(parent.items(), key=operator.itemgetter(1))[1]
    minimum = min(parent.items(), key=operator.itemgetter(1))[1]
    difference = maximum - minimum
    for key1 in temp_dict[key].keys():
      value = temp_dict[key][key1]
      temp_dict[key][key1] = ((value - minimum)/difference)
  return temp_dict
```

```python
with open(root_path + "/query_results_model_video_250.json", 'rb') as video_model:
     video_results = json.load(video_model)
```

```python
with open(root_path + "/query_results_model_video_250.json", 'w') as comb:
     comb.write(json.dumps(video_results))
```

```python
category_dict = {}
for key in video_results.keys():
  category_dict[key] = {}
  result_list = video_results[key]
  for item in result_list:
    value, name, category = item[0],item[1],item[2]
    if name == key:
      continue
    category_dict[key][name] = category
```

```python
video_values = {}
for key in video_results.keys():
  video_values[key] = {}
  result_list = video_results[key]
  for item in result_list:
    value, name, category = item[0],item[1],item[2]
    if name == key:
      continue
    video_values[key][name] = value
```

```python
video_values = normalize(video_values)
```

```python
pprint(video_values)
```

Reading results from text model

```python
with open(root_path + "/query_results_model_text_250.json", 'rb') as text_model:
     text_results = json.load(text_model)
```

```python
temp_dict = {}
for key in text_results.keys():
  temp_dict[key+".mp4"] = {}
  for key1 in text_results[key].keys():
    temp_dict[key+".mp4"][key1 + ".mp4"] = text_results[key][key1]
```

```python
temp_dict = normalize(temp_dict)
```

```python
pprint(temp_dict)
```

combining results

```python
combined_results = {}
for key in temp_dict.keys():
  combined_results[key] = {}
  for key1 in temp_dict[key].keys():
    combined_results[key][key1] = (temp_dict[key][key1] + video_values[key][key1])/2
```

sorting results in descending order for combined

```python
ranked_lists = {}

for key in combined_results.keys():
  ranked_lists[key] = []
  for key1 in combined_results[key].keys():
    ranked_lists[key].append((key1, combined_results[key][key1],  category_dict[key][key1]))
```

```python
ranked_lists.keys()
```

sorting in descending order

```python
for key in ranked_lists.keys():
  result_list = ranked_lists[key]
  result_list = sorted(result_list,key=lambda x: x[1], reverse=True)
  ranked_lists[key] = result_list
```
